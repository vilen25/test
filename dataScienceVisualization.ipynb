```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

#%pip install category_encoders

#from sklearn.preprocessing import LabelBinarizer

#import category_encoders as ce

data_path = "https://raw.githubusercontent.com/ToyeshC/Datasets/main/recruitmentdataset-2022-1.3.csv"
df_data = pd.read_csv(data_path)

df_data.head()
```

```python
fig = plt.figure(figsize=(4,3),dpi=144)
ax = fig.add_subplot(121)
data_A = df_data[df_data['company'] == 'A']
data_B = df_data[df_data['company'] == 'B']
data_C = df_data[df_data['company'] == 'C']
data_D = df_data[df_data['company'] == 'D']
my_labels = ['male', 'female', 'other'] 
ax.pie(data_A[['gender','decision']] [df_data['decision'] == True].value_counts(), labels=my_labels, autopct='%1.1f%%')
plt.title('Percentage of genders that got accepted to company A')
ax = fig.add_subplot(122)
ax.pie(data_A['gender'].value_counts(), labels=my_labels, autopct='%1.1f%%')
plt.title('Percentage of genders that applied to company A', x=0.6)
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=2,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()
#data_A['gender'].value_counts()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
#data_A[['ind-languages','decision']] [df_data['decision'] == True].value_counts()
pd.crosstab(data_A['decision'], data_A['ind-languages']).plot.bar(ax=ax1, stacked=False, title='company A')
pd.crosstab(data_B['decision'], data_B['ind-languages']).plot.bar(ax=ax2, stacked=False, title='Company B')
pd.crosstab(data_C['decision'], data_C['ind-languages']).plot.bar(ax=ax3, stacked=False, title='Company C')
pd.crosstab(data_D['decision'], data_D['ind-languages']).plot.bar(ax=ax4, stacked=False, title='Company D')
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=2,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)

ax1.hist(data_A['age'] [data_A['decision'] == True], alpha=0.8, label='True decision')
ax1.hist(data_A['age'], alpha=0.5, label='Total applied')
ax1.legend(loc='upper right')
ax1.set_title('Company A')

ax2.hist(data_B['age'] [data_B['decision'] == True], alpha=0.8, label='True decision')
ax2.hist(data_B['age'], alpha=0.5, label='Total applied')
ax2.legend(loc='upper right')
ax2.set_title('Company B')

ax3.hist(data_C['age'] [data_C['decision'] == True], alpha=0.8, label='True decision')
ax3.hist(data_C['age'], alpha=0.5, label='Total applied')
ax3.legend(loc='upper right')
ax3.set_title('Company C')

ax4.hist(data_D['age'] [data_D['decision'] == True], alpha=0.8, label='True decision')
ax4.hist(data_D['age'], alpha=0.5, label='Total applied')
ax4.legend(loc='upper right')
ax4.set_title('Company D')

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=2,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()
```
